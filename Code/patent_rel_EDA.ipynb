当然，以下是这些数据字段的中文解释：

1. **patent_num**: 专利编号
2. **cpc**: 国际专利分类 (Cooperative Patent Classification)
3. **cpc_1**: CPC 一级分类代码
4. **cpc_2**: CPC 二级分类代码
5. **cpc_g1**: CPC 一级分类的数量组
6. **cpc_g2**: CPC 二级分类的数量组
7. **cpc_3**: CPC 三级分类代码
8. **cpc_g3**: CPC 三级分类的数量组
9. **permno**: 永久编号
10. **issue_date**: 授权日期
11. **filing_date**: 申请日期
12. **xi_nominal**: 名义价值指数
13. **xi_real**: 实际价值指数
14. **cites**: 被引用次数
15. **year_issue**: 授权年份
16. **month_issue**: 授权月份
17. **year_filing**: 申请年份
18. **month_filing**: 申请月份
